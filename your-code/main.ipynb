#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
sns.set(style='white',color_codes=True)
sns.boxplot(x='Age',data=titanic,palette='pastel').set_title("Age of passengers",weight='bold')
sns.despine(trim=True) 
```

#### What does the last plot tell you about the feature `Age`?

The bulk of passengers were under 40 years-old. People over 60 years-old are considered outliers. This is consistent with the life expectancy in UK at the time (between 50 and 55 years-old depending on the gender).

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.set(style='white',color_codes=True)
sns.violinplot(y='Age',inner='quartile',data=titanic,palette='rainbow').set_title("Distribution of age",fontsize=13)
sns.despine(trim=True) 
```

#### What additional information does the last plot provide about feature `Age`?

In addition of the boxplot, it also shows us the distribution of the age. This distribution is slightly right-skewed. Once again this is consistent with the life expectancy at the time.

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib


plt.scatter(x='Age',y='Fare',data=titanic)
plt.legend()
```

```python
# Method 2 - seaborn

fig, axs = plt.subplots(1,2,sharex=True)
sns.relplot(x='Age',y='Fare',data=titanic,ax=axs[0])
sns.relplot(x='Age',y='Fare',data=titanic,kind='line',ax=axs[1])
```

#### Plot the correlation matrix using seaborn.

```python
sns.set(color_codes = False)
corr = titanic.corr()
mask_ut=np.triu(np.ones(corr.shape)).astype(np.bool)

fig, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(corr,mask=mask_ut, cmap = cmap, center = 0, annot = True)

plt.show()
```

#### What are the most correlated features?

Passenger class and Fare, which is logic. 
Passenger class and Age, which can be understandable as well.
